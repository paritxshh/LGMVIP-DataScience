# IRIS Flower Classification ML Project

### Python machine learning tasks that are well-known and enduring include classifying the iris flower dataset. 150 samples of iris blossoms from one of three species: Setosa, Versicolor or Virginica are included in the dataset. The objective is to create a model that can correctly categorize the species using the supplied features. Python and certain well-known libraries such as scikit-learn and matplotlib, can be used to do this.
### Here is a detailed explanation on how to accomplish it:-

## 1. Importing the necessary libraries 

```python
#pandas helps us to load the data.
import pandas as pd

#numpy is used for computational operations.
import numpy as np

#matplotlib and seaborn are used for data visualization.
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## 2. Loading the data

```python
from sklearn.preprocessing import LabelEncoder  #converts categorical values into numerical form.
le=LabelEncoder()
df["Class_Labels"]=le.fit_transform(df["Class_Labels"])
```

```python
x=df.iloc[:,:4]
y=df.iloc[:,4]
```

## 4. Splitting the Data into Training and Testing datasets.

### We must divide the entire dataset into train and test sets in order to first test the model and then train it. In such case, the test dataset is used to test the model that has been trained using the training dataset. The training dataset is used to train the model.

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=0)
```

## 5. Building the Model

### We will use logistic regression to categorise the Iris flowers dataset in this instance. 
### Import LogisticRegression from the sci-kit learn package to start building the model.

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
y_pred
```

### Finding the Accuracy and Confusion Matrix

### The accuracy score indicates the predictive power of the model we develop, and the confusion matrix contains a matrix of actual and anticipated values. Import confusion_matrix and accuracy_score from the sci-kit learn metric package to do so.

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
confusion_matrix(y_test,y_pred)
```

```python
accuracy=accuracy_score(y_test,y_pred)*100
print("Accuracy of the model is {:.2f}".format(accuracy))
```

## 6. Conclusion

### Setosa, Versicolor, and Virginica are the three distinct species of iris blooms that can be distinguished using the Iris dataset classification using logistic regression. 
### For this objective, logistic regression is a common and straightforward approach that performs well with binary and multi-class classification issues.
### Throughout the analysis, by training the logistic regression model on a subset of the dataset and assessing its performance on the remaining unobserved data, we have noticed encouraging results in properly categorising the iris flowers.
### And we obtained an accuracy of 97.37%, demonstrating the high precision of the model we developed.
